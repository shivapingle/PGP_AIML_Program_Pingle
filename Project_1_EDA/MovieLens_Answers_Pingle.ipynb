```python
#row 267 is missing
new_267 = {'movie id':267,'movie title': 'TRIAL MOVIE','release date':'01-Jan-1995','Comedy':1}
item_df2_trial = item_df2.append(new_267, ignore_index=True)
item_df2_trial.fillna(value=0.0,inplace=True)
item_df2_trial.sort_values(by=['movie id'],inplace=True)
item_df2_trial = item_df2_trial.reset_index()
item_df2_trial[264:270]
```

```python
aa = data_df1.groupby('movie id',as_index=False).count()
movie_list_100 = aa[aa['rating']>100]
#movie_list_100.reset_index(inplace=True)
movie_list_100.drop(['user id','timestamp'],axis=1,inplace=True)
```

```python
bb = data_df1.groupby('movie id',as_index=False).mean()
#bb.reset_index(inplace=True)
rating_descending = bb.sort_values(by=['rating'],ascending=False)
rating_descending.drop(['timestamp'],axis=1,inplace=True)
#dd = rating_descending.reset_index()
#dd.head(10)
```

```python
Selected_top_25 = rating_descending[rating_descending['movie id'].isin(movie_list_100['movie id'])][0:25]
Selected_top_25.drop(['user id'],axis=1,inplace=True)
Selected_top_25.sort_values(by=['movie id'],inplace=True)
```

```python
xx = item_df2[item_df2['movie id'].isin(Selected_top_25['movie id'])]
Top_25 = pd.merge(Selected_top_25,xx[['movie id','movie title']],how='inner',on='movie id')
Top_25.sort_values(by=['rating'],inplace=True,ascending=False)
print(f"The list of top 25 movies by rating:\n{Top_25}")
```

```python
Genre_by_gender.div(Genre_by_gender.total, axis= 0) * 100  #dividing each cell with row total and multiplying by 100
```
